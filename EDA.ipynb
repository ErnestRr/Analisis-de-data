Análisis Exploratorio de Datos

El Análisis Exploratorio de Datos o EDA es la tarea más importante para un Científico de Datos, pues constituye el primer paso antes de preparar algún modelo.

Cuando se recibe la información por primera vez, es importante comprender el conjunto de datos, así como reconocer patrones y distribuciones que puedan ser útiles para entender el comportamiento de las variables e identificar anomalías que puedan afectar al momento de realizar el modelo.

A veces, cuando se tiene un conjunto de datos y se pretende hacer un análisis exploratorio, el punto de inicio puede dificultarse; sin embargo, existen algunas preguntas que podrían ser útiles, tales como las siguientes:

¿Cuántos registros (o filas con datos) hay?
¿Se tienen muchas o pocas observaciones?
¿Los datos están completos o falta información?
¿Qué tipo de datos son, continuos y/o discretos (esto se explicará más adelante?
¿Se pueden identificar los tipos de datos de cada columna?
¿Los datos están balanceados de acuerdo con el objetivo?
¿Cuáles son elementos o columnas importantes? ¿Cuáles se pueden descartar?
¿Las variables siguen alguna distribución estadística conocida?
¿Hay correlación entre las variables?
¿Existen datos raros o atípicos en el conjunto de datos?
En ocasiones, se tiene un conjunto de datos incompleto o insuficiente; ante eso, un análisis exploratorio permitirá identificar si son viables o no para realizar algún modelo.

Completitud
La completitud mide cuánta información se tiene sobre una variable; en otras palabras, indica los datos que están completos en el conjunto de datos. Para saberlo, se realiza una tabla con una nueva variable. Su aplicación en el ejemplo dfsería como sigue:

Se define una variable denominada completitud, semejante al conjunto de datos df. En este caso, se utilizará la función .sum()) para sumar los elementos que están vacíos (.isnull()). Al juntar estas tres caractarísticas se obtiene df.isnull().sum() y dado que se pretende hacer un Data Frame ,entonces se usará pd.DataFrame.

```python
#Verificamos completitud de las variables , es decir,  si tenemos valores nulos
completitud = pd.DataFrame(df.isnull().sum())
completitud
```

Como resultado, para cada variable (o columna) de df se determina cuántos valores distintos de cero tiene. De esta manera, a mayor cantidad mostrada, mayor número de datos faltantes.

De lo anterior, resultará un data frame donde cada «fila» constituye una de las columnas del df original; como se busca que la primera fila sea la 0, la segunda 1 y así sucesivamente, entonces se usará reset_indexpara resetear los índices. El argumento inplace= True sirve para evitar la repetición de completitud =completitud.reset_index(), aunque también es válido.

```python
completitud.reset_index(inplace = True)
completitud
```

```python
#Se crean encabezados
completitud=completitud.rename(columns = {"index":"columna",0:"total"})
completitud
```

Se creará una nueva columna llamada "completitud", igual a 1 - la división de la columna "total" entre la cantidad de columnas del conjunto de datos original. El resultado se multiplica por 100 para obtener un procentaje.

Completitudi=1−totalitotal columnas×100

```python
completitud["completitud"] = (1 - completitud["total"] / df.shape[0] * 100)
completitud
```

Por último, se organiza de mayor a menor con .sort_values (no olvides indicar ascending=True para obtener el orden correcto) y se resetean nuevamente los índices.

```python
completitud = completitud.sort_values(by = "completitud", ascending = True)
completitud.reset_index(drop = True, inplace = True)
completitud
```

```python
#Creamos una función que nos permita modificar el nombre de nuestras columnas
def rename_cols(df,cols,prefix):
    new_feats=[prefix+col for col in cols]
    df=df.rename(columns=dict(zip(cols,new_feats)))
    return df
```

Como verás, ya tienes tres variables nuevas que incluyen el nombre de todas las columnas y están asignadas, ya sea a v_feats, c_feats o tgt.

Esta parte se hace de forma manual y viendo la naturaleza de cada columna.

Ahora, puedes aplicar la función rename_cols creada con anterioridad para modificar el nombre de las columnas a uno más comprensible.

Análisis Univariable
Luego de detectar los tipos de variables presentes en el conjunto de datos, es posible identificar una en especial: la variable objetivo.

Al identificar esta variable objetivo, se aplica el análisis univariable, basado en estudiar la variable objetivo y su relación con las demás variables para encontrar correlaciones o algún dato importante relacionado con esta variable.

```python
#La sentencia "describe" nos indica la estadistica
df.filter(like="tgt_").describe()
```

```python
sns.distplot(df['tgt_SalePrice']);
```

La gráfica de arriba es una histograma e indica cómo se comportan todos los números que están en la variable objetivo, llamado tgt_SalePrice.

Se observa que la mayoría de los resultados está por debajo de 200 000 pero por encima de 100 000 y que los valores arriba de 300 000 son muy pocos.

Este tipo de comportamiento se llama distribución normal,es decir, la mayoría de los números se concentra en un punto específico y se disipa a lo largo de la recta.

Asimetría Medida de distribución que puede tomar valores positivos o negativos. En caso de ser positivo, indica que la distribución se inclina hacia los valores positivos; en caso contrario, los valores negativos muestran que la distribución se inclina hacia los valores negativos.

Curtosis La curtosis es una medida estadística que determina el grado de acumulación en los valores de una variable alrededor del centro de la distribución. Cuando tiende a 3, significa que sigue una distribución normal, es decir, que los valores no son atípicos.

```python
print("Asimetría: %f" % df['tgt_SalePrice'].skew())
print("Curtosis: %f" % df['tgt_SalePrice'].kurt())
```

Análisis multivariante
Anteriormente, conociste el "análisis univariable", que sucede entre cada variable con la variable objetivo. Sin embargo, para comparar múltiples variables contra otras tantas (más allá del objetivo), entonces se efectúa un "análisis multivariable".

Este análisis permitirá entender la relación de las variables entre sí para detectar anomalías o patrones internos.

```python

# Selecciona solo las columnas numéricas
numeric_df = df.select_dtypes(include=[np.number])
```

```python
#Matriz de correlación general entre todas las variables
corr = numeric_df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

El mapa de calor permite el conocimiento visual de las variables y sus relaciones; cuando están muy correlacionadas podría indicar multicolinealidad, esto es, que tienen la misma información y no serían útiles para nuestro modelo.

Esta gráfica proveniene del "análisis univariable", en particular, del estudio de la correlación entre cada una de las variables y la variable objetivo; sin embargo, en este otro caso se comparan las correlaciones de todas las variables contra todas las variables.

Esto genera una matriz para señalar los valores que, de entre los distintos factores dentro un conjunto de datos, se comportan de manera similar o no tienen relación alguna.

Aunque para este momento no resulte del todo intelegible, en cuanto avances con los modelos, estos cobrarán sentido e importancia.

```python
#En vista de nuestras variables más correlacionadas graficaremos un "Scatter plot" para facilitar la comprensión de nuestras variables
sns.set()
cols = ['tgt_SalePrice', 'v_OverallQual', 'c_GrLivArea', 'c_GarageCars', 'c_TotalBsmtSF', 'v_FullBath']
sns.pairplot(df[cols], size = 2.5)
plt.show();
```

La gráfica anterior se asemeja al mapa de calor previo, la diferencia radica en que muestra pequeñas gráficas de la relación de todas las variables contra todas las variables, así, se entiende como la forma "gráfica" de corroborar si existe correlación.

Recuerda que si observas que una línea se dibuja quiere decir que los valores de esas variables poseen una alta correlación.
